```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
%matplotlib inline

chengdu = mpimg.imread('challenge-7-chengdu.png') 
plt.imshow(chengdu) # 
```

```python
chengdu.shape
```

```python
data = chengdu.reshape(516*819,3)
data.shape,data[10]
```

```python
def get_variety(data):
    temp = data.tolist()
    variety = len(set([tuple(t) for t in temp]))
    return variety
get_variety(data),data[20]
```

```python
from sklearn.cluster import MiniBatchKMeans

model = MiniBatchKMeans(n_clusters=25)
model.fit(data)
predit = model.predict(data)
new_colors = model.cluster_centers_[predit]
```

```python
get_variety(new_colors)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

new_chengdu = new_colors.reshape(chengdu.shape)
ax[0].imshow(chengdu)
ax[1].imshow(new_chengdu)
```

```python
mpimg.imsave("new_chengdu.png", new_chengdu)
```
