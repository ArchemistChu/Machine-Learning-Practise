```python
import pandas as pd
df = pd.read_csv('challenge-8-seeds.csv')
df.head()
```

```python
from scipy.cluster import hierarchy
H = hierarchy.linkage(df,'ward','euclidean') # method:ward, metric:euclidean
```

```python
H[:5]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.figure(figsize=(15, 8))
hierarchy.dendrogram(H)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
hierarchy.dendrogram(H, truncate_mode='lastp', p=16, show_leaf_counts=True, show_contracted=True)
plt.show()
```

The data should be clustered to 3 categories, which means contains 3 different types of seeds
