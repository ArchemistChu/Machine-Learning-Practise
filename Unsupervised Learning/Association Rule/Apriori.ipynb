$$
SUPPORT(A \to B) = \frac{\text{(Transaction contains A and B) } }{\text{Total number of transactions} }
$$

$$
CONFIDENCE(A \to B) = \frac{\text{(Transaction contains A and B) } }{\text{Number of transactions containing A } }
$$

For an itemset contains n items, the potential rules are:
$$
N = 3^n - 2^{n+1} + 1
$$

```python
dataset = [
    ["Bread", "Milk", "Butter"],
    ["Bread", "Diapers", "Milk", "Beer"],
    ["Milk", "Diapers", "Beer", "Orange Juice"],
    ["Bread", "Milk", "Diapers", "Butter"],
    ["Bread", "Beer", "Orange Juice"],
    ["Milk", "Butter"],
    ["Bread", "Milk", "Beer", "Butter"],
    ["Diapers", "Orange Juice"],
    ["Bread", "Milk", "Diapers", "Beer", "Butter"],
    ["Bread", "Diapers", "Orange Juice", "Butter"],
    ["Eggs", "Cereal", "Milk"],
    ["Bread", "Eggs", "Cereal", "Orange Juice"]
]
dataset
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

Te = TransactionEncoder()
Te_ary = Te.fit(dataset).transform(dataset)
Te_ary
```

```python
df = pd.DataFrame(Te_ary, columns=Te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.3, use_colnames=True)
```

```python
freq_itemsets = apriori(df, min_support=0.3, use_colnames=True)
freq_itemsets[freq_itemsets.itemsets.apply(lambda x: len(x)) >= 2]
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(freq_itemsets,20, metric="confidence", min_threshold=0.6)
```

```python
association_rules(freq_itemsets,20, metric="confidence", min_threshold=0.8)
```

$$
lift(A \to B) = \frac{support(A \cap B)}{support(A) \times support(B)}
$$

When 
The larger the lift value, the stronger the correlation between the antecedent and the consequent

$$
leverage(A \to B) = support(A \cap B) - support(A) \times support(B)

The larger the leverage value, the stronger the correlation between the antecedent and the consequent

$$
conviction(A \to B) = \frac{1 - support(B)}{1 - confidence(A \to B)}
$$

The larger the convicion value, the stronger the correlation between the antecedent and the consequent
